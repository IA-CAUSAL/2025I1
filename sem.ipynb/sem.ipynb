```python
def codificar_codigo_ens(value):
    value=int(value)
    if value==310:
        return 0
    elif value>=360 and value<=363:
        return 0
    else:
        return 1
```

```python
data["ENS_TYPE"]=data["ENS_CODE"].apply(codificar_codigo_ens)
```

```python
import semopy
from semopy import Model
```

```python
grafo = """
HOW_MANY_WORK_FAMILY_GROUP ~ PER_CAPITA_HOUSEHOLD_INCOME + PEOPLE_STUDYING_HIGHER + MOTHER_EDUCATION + FATHER_EDUCATION
HAS_WORK_REM ~ STUDY_HIGH_INSTITUTION + ENS_TYPE
SCORE_NEM ~ AVG_CM_CURRENT + HOW_MANY_WORK_FAMILY_GROUP  + ENS_TYPE + HAS_WORK_REM 
"""
model=Model(grafo)
res_opt = model.fit(pd.DataFrame(x,columns=features),obj="DWLS")
estimates = model.inspect()
g = semopy.semplot(model, "sem.png")
stats = semopy.calc_stats(model)
print(stats)
```
